We can confirm what the documentation states:

```python
for column in data.columns:
    count = data[column].value_counts().to_frame().sort_index(ascending=True)
    if '?' in count.index:
        print("{0}: {1}".format(count.loc["?"].index[0], count.loc["?"][0]))
```

We will temporarily substitute '?', aka missing values, with -1 to make the values easier chart and also because DecisionTreeClassifier doesn't support missing values:

```python
import matplotlib.pyplot as plt

data_no_q = data.replace('?', -1)

for column in data_no_q.columns:
    fig = plt.figure()
    pd.value_counts(pd.to_numeric(data_no_q[column])).sort_index(ascending=True).plot(kind='bar', figsize=(8, 4), title=column)
    fig.canvas.draw()
```

```python
from sklearn import tree, model_selection

rseed=100
data_train, data_test = model_selection.train_test_split(data_no_q, random_state=rseed, test_size=0.1)

train_labels = data_train['CLASS'].values
train_values = data_train.drop(['CLASS'], axis=1)
test_labels = data_test['CLASS'].values
test_values = data_test.drop(['CLASS'], axis=1)

clf = tree.DecisionTreeClassifier()
clf.fit(train_values, train_labels)
clf.score(test_values, test_labels)
```

```python
import pydotplus
from IPython.display import Image  

dot_data = tree.export_graphviz(clf, out_file=None, 
                         feature_names=names.as_matrix()[0],
                         class_names=['DIE', 'LIVE'],  
                         filled=True, rounded=True,  
                         special_characters=True)  
graph = pydotplus.graph_from_dot_data(dot_data)  
Image(graph.create_png())  
```

```python
import numpy as np

# Following descretization bins set in the documentation...
age_bin = list(range(10, 90, 10))
bilirubin_bin = (0, 0.39, 0.80, 1.20, 2.00, 3.00, 4.00)
alk_phos_bin = (0, 33, 80, 120, 160, 200, 250)
sgot_bin = (0, 13, 100, 200, 300, 400, 500)
albumin_bin = (0, 2.1, 3.0, 3.8, 4.5, 5.0, 6.0)
protime_bin = list(range(0, 100, 10))

data_no_q['AGE'] = np.digitize(data_no_q['AGE'], age_bin)
data_no_q['BILIRUBIN'] = np.digitize(data_no_q['BILIRUBIN'].astype(float), bilirubin_bin)
data_no_q['ALKPHOSPHATE'] = np.digitize(data_no_q['ALKPHOSPHATE'].astype(float), alk_phos_bin)
data_no_q['SGOT'] = np.digitize(data_no_q['SGOT'].astype(float), sgot_bin)
data_no_q['ALBUMIN'] = np.digitize(data_no_q['ALBUMIN'].astype(float), albumin_bin)
data_no_q['PROTIME'] = np.digitize(data_no_q['PROTIME'].astype(float), protime_bin)
```

```python
for column in data_no_q.columns:
    fig = plt.figure()
    pd.value_counts(pd.to_numeric(data_no_q[column])).sort_index(ascending=True).plot(kind='bar', figsize=(8, 4), title=column)
    fig.canvas.draw()
```

```python
rseed=100
data_train, data_test = model_selection.train_test_split(data_no_q, random_state=rseed, test_size=0.1)

train_labels = data_train['CLASS'].values
train_values = data_train.drop(['CLASS'], axis=1)
test_labels = data_test['CLASS'].values
test_values = data_test.drop(['CLASS'], axis=1)

clf2 = tree.DecisionTreeClassifier()
clf2.fit(train_values, train_labels)
clf2.score(test_values, test_labels)
```

```python
dot_data = tree.export_graphviz(clf2, out_file=None, 
                         feature_names=names.as_matrix()[0],
                         class_names=['DIE', 'LIVE'],  
                         filled=True, rounded=True,  
                         special_characters=True)  
graph = pydotplus.graph_from_dot_data(dot_data)  
Image(graph.create_png())  
```
