```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns

fileURL = 'http://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
data = pd.read_csv(fileURL, names=['Sepal Length', 'Sepal Width', 'Petal Length', 'Petal Width', 'Name'], header=None)
```

```python
data.head()
```

There are 5 features - Sepal Length, Sepal Width, Petal Length, Petal Width, and Name.  Sepal Length, Sepal Width, Petal Length, and Petal Width are discrete numeric values.  Name is nominal. 

```python
num_data = data.drop("Name", 1)
pd.DataFrame([num_data.min(axis=0), num_data.max(axis=0), num_data.mean(axis=0), num_data.max(axis=0)-num_data.min(axis=0), num_data.std(axis=0), num_data.var(axis=0), num_data.count(axis=0), num_data.quantile(0.25), num_data.quantile(0.5), num_data.quantile(0.75)], index=["Min", "Max", "Mean", "Range", "Std-dev", "Variance", "Count", "25%", "50%", "75%"])
```

```python
%matplotlib inline

ax = data.plot.hist(alpha=0.5)
data.plot.hist(subplots=True, layout=(2, 2))
data.plot.box(notch=True)
```

```python
import pandas as pd
fileURL = 'http://archive.ics.uci.edu/ml/machine-learning-databases/pendigits/pendigits.tra'
digits = pd.read_csv(fileURL, header=None)
```

```python
digits.head()
```

There are 17 features - features 0-15 are numeric data and feature 16 is nominal (class).

```python
pd.DataFrame([digits.min(axis=0), digits.max(axis=0), digits.mean(axis=0), digits.max(axis=0)-digits.min(axis=0), digits.std(axis=0), digits.var(axis=0), digits.count(axis=0), digits.quantile(0.25), digits.quantile(0.5), digits.quantile(0.75)], index=["Min", "Max", "Mean", "Range", "Std-dev", "Variance", "Count", "25%", "50%", "75%"])
```

```python
digits.plot.hist(subplots=True, figsize=(10, 10), layout=(6, 3))
```

```python
#make it a little larger since it's hard to see
fig, ax = plt.subplots()
fig.set_size_inches(18.5,10.5)
digits.boxplot(ax=ax, showfliers=True, sym='k.', notch=True)
plt.show()
```

---------------------------------

Sepal length is more concentrated at a median than petal length and petal length has a group of low values that fall out of an expected single distribution.  Likewise, sepal width is at a higher median than petal width, and sepal length has a higher peak value than petal width's more block-shaped distribution.  Petal length could be separated at around 3 to create two parts.

Sepal length and petal width appear to have significantly different medians.  Petal length appears to explain the greatest amount of the data.

Features 1 and 3 have outliers, while all other features except 16 have whiskers set at the limits of the range of that feature, implying there are no outliers for those features.  Feature 16 has no outliers because it's a class feature.

The second and forth histograms have what appears to be something akin to an exponential distribution, with high frequency of values clustered at the higher end with a trailing end of lower values.  This would explain the outliers in these features because the mean is very high with a smaller standard deviation, so many outliers can exist.
